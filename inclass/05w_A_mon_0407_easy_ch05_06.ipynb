```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'var1' : [4, 3, 8],
                   'var2' : [2, 6, 1]})
df
```

```python
df1=df.copy()
df1
```

```python
df1['var_sum'] = df1['var1'] + df1['var2']
df1
```

```python
df1["var_mean"] = (df1["var1"] + df1["var2"]) / 2
df1
```
